```python

from tensorflow.keras.preprocessing.text import one_hot # pyright: ignore[reportMissingImports]
```

```python
### sentences
sent=[  'the glass of milk',
    'the glass of juice',
    'the cup of tea',
    'I am a good boy',
    'I am a good developer',
    'understand the meaning of words',
    'your videos are good',]
sent 
```

```python

# vocabulary size
vocab_size=10000
```

```python
## One Hot Representation
onehot_repr=[one_hot(words,vocab_size)for words in sent]
print(onehot_repr)
```

```python
## word Embedding Representation
from tensorflow.keras.layers import Embedding  # pyright: ignore[reportMissingImports]
from tensorflow.keras.preprocessing.sequence import pad_sequences  # pyright: ignore[reportMissingImports]
from tensorflow.keras.models import Sequential  # pyright: ignore[reportMissingImports]
```

```python
import numpy as np
```

```python
sent_length=8
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)
```

```python
#feature representation
dim=10
```

```python
model=Sequential()
model.add(Embedding(vocab_size,dim,input_length=sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
